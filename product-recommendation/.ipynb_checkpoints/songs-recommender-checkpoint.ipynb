```python
import turicreate as tc
import turicreate.aggregate as agg
```

# Load Data

```python
song_data = tc.SFrame('song_data.sframe')
```

# Explore Data

### Shape of Data

```python
song_data.head()
```

```python
### Most frequently played songs
```

```python
listen_freq = song_data.groupby(key_column_names='song', operations={'count': agg.SUM('listen_count')})
listen_freq = listen_freq.sort('count', ascending=False)
listen_freq
```

### Number of songs

```python
len(listen_freq)
```

```python
songs = song_data['song'].unique()
len(songs)
```

### Number of users

```python
len(song_data.groupby(key_column_names='user_id', operations={'count': agg.COUNT()}))
```

```python
users = song_data['user_id'].unique()
len(users)
```

### Artists that have most unique users listening to their songs

```python
pop_artists = song_data.groupby(key_column_names='artist', operations={'count': agg.COUNT_DISTINCT('user_id')})
pop_artists = pop_artists.sort('count', ascending=False)
pop_artists[:10]
```

```python
pop_artists.filter_by(['Kanye West', 'Foo Fighters', 'Taylor Swift', 'Lady GaGa'], 'artist')
```

### Artists with the most total listen_count

```python
pop_artists = song_data.groupby(key_column_names='artist', operations={'count': agg.SUM('listen_count')})
pop_artists = pop_artists.sort('count', ascending=False)
pop_artists[:10]
```

```python
pop_artists.filter_by(['Kings Of Leon', 'Coldplay', 'Taylor Swift', 'Lady GaGa'], 'artist')
```

```python
pop_artists.filter_by(['William Tabbert', 'Velvet Underground & Nico', 'Kanye West', 'The Cool Kids'], 'artist')
```

# Create a song recommender

```python
train_data, test_data = song_data.random_split(.8, seed=0)
```

### Simple popularity-based recommender

```python
pop_model = tc.popularity_recommender.create(train_data,
                                             user_id='user_id',
                                             item_id='song')
```

### Use the popularity model to make some predictions

```python
pop_model.recommend(users=[users[0]])
```

```python
pop_model.recommend(users=[users[1]])
```

# Build a personalized recommender

```python
person_model = tc.item_similarity_recommender.create(train_data,
                                                     user_id='user_id',
                                                     item_id='song')
```

### User personalized model to make some predictions

```python
person_model.recommend(users=[users[0]])
```

```python
person_model.recommend(users=[users[1]])
```

```python
person_model.get_similar_items(['With Or Without You - U2'])
```

```python
person_model.get_similar_items(['Chan Chan (Live) - Buena Vista Social Club'])
```

# Quantitative comparison between models

```python
model_performance = tc.recommender.util.compare_models(test_data, 
                                                       [pop_model, person_model], 
                                                       user_sample=.05)
```

```python
x = model_performance[0]['precision_recall_overall']['recall']
y = model_performance[0]['precision_recall_overall']['precision']
tc.visualization.scatter(x, y)
```

```python
x = model_performance[1]['precision_recall_overall']['recall']
y = model_performance[1]['precision_recall_overall']['precision']
tc.visualization.scatter(x, y)
```
